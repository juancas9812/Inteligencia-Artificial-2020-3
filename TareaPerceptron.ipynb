# 2. Implementación del perceptrón en sklearn

Se crea un dataset en espiral similar al de tensorflow, y se dividen los datos en train y test en un ratio de 70-30.

Posteriormente, se crea el clasificador con 3 capas internas, la primera de 6 neuronas, la segunda de 5 y la tercera de 4, y la activación es la ReLU.

```
from sklearn.datasets import make_classification, make_moons,make_circles
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score

def make_spirals(n_points, noise=.1):
    n = np.sqrt(np.random.rand(n_points,1)) * 550 * (2*np.pi)/360
    d1x = -np.cos(n)*n + np.random.rand(n_points,1) * noise
    d1y = np.sin(n)*n + np.random.rand(n_points,1) * noise
    return (np.vstack((np.hstack((d1x,d1y)),np.hstack((-d1x,-d1y)))), 
            np.hstack((np.zeros(n_points),np.ones(n_points))))


X, y = make_spirals(1000)
plt.title('training set')
plt.plot(X[y==0,0], X[y==0,1], '.', label='class 1')
plt.plot(X[y==1,0], X[y==1,1], '.', label='class 2')
plt.legend()
plt.show()

X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=1, test_size=0.3)

clf = MLPClassifier(hidden_layer_sizes=(6, 5, 4),activation='relu',random_state=1, max_iter=2000,verbose=True)
clf.fit(X_train, y_train)
y_ANN_predict=clf.predict(X_test)
ACC = accuracy_score(y_test, y_ANN_predict)
print("ACC:", ACC*100)
```

```
clf2 = MLPClassifier(hidden_layer_sizes=(5,4, 3),activation='relu',random_state=1, max_iter=2000,verbose=True)
clf2.fit(X_train, y_train)
y_ANN_predict2=clf2.predict(X_test)
ACC = accuracy_score(y_test, y_ANN_predict2)
print("ACC:", ACC*100)
```
